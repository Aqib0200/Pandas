```python
price_map = {'Software': 350, 'Hardware': 425, 'Service': 275}
df['Revenue'] = df['Units'] * df['Product'].map(price_map)
display(df.head())
```

```python

df['DayOfWeek'] = df.index.day_name()
display(df[['DayOfWeek']].head())
```

```python
def apply_discount(row):
    units = row['Units']
    revenue = row['Revenue']
    if units >= 20:
        discount = 0.10
    elif units >= 15:
        discount = 0.05
    else:
        discount = 0.0
    return revenue * (1 - discount), discount

df[['RevenueAfterDiscount', 'DiscountPct']] = df.apply(
    lambda r: pd.Series(apply_discount(r)), axis=1
)

# Make DiscountPct human-readable (optional)
df['DiscountPct'] = df['DiscountPct'] * 100   # percentages
display(df[['Units','Revenue','DiscountPct','RevenueAfterDiscount']].head())
```

```python
df['Company'] = df['Company'].str.upper()
df['CompanyCode'] = df['Company'].str[:3]
display(df[['Company','CompanyCode']].drop_duplicates().head())
```
